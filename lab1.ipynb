# 8. Кто из мужчин является ветераном компании?

```python
df["Start Date"] = pn.to_datetime(df["Start Date"])
veteran = df[df["Gender"] == "Male"].loc[df[df["Gender"] == "Male"]["Start Date"].idxmin()]
veteran
```

# 9. Отсортируйте служащих по убыванию заработной платы и найдите пять человек, получающих наибольшую заработную плату

```python
top_salary = df.sort_values("Salary", ascending=False).head(5)
print(top_salary)
```

# 10.Определите наиболее распространенное в компании мужское и женское имя

```python
most_common_male = df[df["Gender"] == "Male"]["First Name"].mode()[0]
most_common_female = df[df["Gender"] == "Female"]["First Name"].mode()[0]

print(most_common_male)
print(most_common_female)
```

# 11. Какой процент служащих относится к высшему руководству?

```python
senior_pct = (df["Senior Management"].sum() / len(df)) * 100
print(senior_pct)
```

# 12. Определите пять служащих, входивших в систему самыми последними

```python
df["Last Login Time"] = pn.to_datetime(df["Last Login Time"])
latest_logins = df.sort_values("Last Login Time", ascending=False).head(5)
print(latest_logins)
```

# 13.Имеет ли место гендерное неравенство в отношении заработной платы и бонусов?

```python
gender_stats = df.groupby("Gender")[["Salary", "Bonus %"]].mean()
gender_stats
```

# 14. В каком отделе работает больше всего женщин?

```python
top_women_team = df[df["Gender"] == "Female"]["Team"].mode()[0]
top_women_team
```

# 15. Какова средняя заработная плата и средние бонусы в компании?

```python
avg_salary = df["Salary"].mean() #Cредняя зп
avg_bonus = df["Bonus %"].mean() #Средний бонус

print(avg_salary)
print(avg_bonus)
```
